```python
import numpy as np
from scipy.special import softmax
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import numpy as np

base_path = "/kaggle/input/parkinson-modal-predictions-for-fusion/"

voice_probs = np.load(base_path + 'pred_probs_voice.npy')
voice_labels = np.load(base_path + 'true_labels_voice.npy')

hand_probs  = np.load(base_path + 'pred_probs_hand.npy')
hand_labels = np.load(base_path + 'true_labels_hand.npy')

gait_probs  = np.load(base_path + 'pred_probs_gait.npy')
gait_labels = np.load(base_path + 'true_labels_gait.npy')

```

```python
# === Match lengths across modalities ===
n = min(len(voice_probs), len(hand_probs), len(gait_probs))
voice_probs  = voice_probs[:n]
hand_probs   = hand_probs[:n]
gait_probs   = gait_probs[:n] 
true_labels  = true_labels = voice_labels[:n]  # using voice labels as ground truth 

# === Softmax to normalize (recommended for numerical stability) ===
voice_probs = softmax(voice_probs, axis=1)
hand_probs  = softmax(hand_probs, axis=1)
gait_probs  = softmax(gait_probs, axis=1)
```

```python
# === Late fusion (average the probabilities) ===
fused_probs = (voice_probs + hand_probs + gait_probs) / 3
fused_preds = np.argmax(fused_probs, axis=1)
```

```python
# === Classification Metrics ===
acc = accuracy_score(true_labels, fused_preds)
f1 = f1_score(true_labels, fused_preds, average='weighted')

print(f"✅ Fused Accuracy: {acc:.4f}")
print(f"✅ Fused F1 Score: {f1:.4f}")
```
