```python
import tensorflow as tf
import numpy as np
import cv2

# --- CONFIG ---
MODEL_PATH = '/kaggle/input/spiral-wave/best_dual_model.h5'  # path to your model
SPIRAL_IMG_PATH = '/kaggle/input/test-my-hw/Test_my_hw_sp.jpg'
WAVE_IMG_PATH = '/kaggle/input/test-my-hw/Test_my_hw_wave.jpg'
IMG_SIZE = (256, 256)

# --- LOAD MODEL ---
model = tf.keras.models.load_model(MODEL_PATH)

# --- PREPROCESS FUNCTION ---
def preprocess_image(path, img_size=(256, 256)):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    img = 255 - img  # invert, if your training did this
    img = cv2.resize(img, img_size)
    img = img.astype(np.float32) / 255.0
    img = np.expand_dims(img, axis=-1)  # add channel dimension
    img = np.expand_dims(img, axis=0)   # add batch dimension
    return img

# --- LOAD IMAGES ---
spiral_input = preprocess_image(SPIRAL_IMG_PATH)
wave_input = preprocess_image(WAVE_IMG_PATH)

# --- PREDICT ---
prediction = model.predict({'spiral': spiral_input, 'wave': wave_input})[0][0]

# --- INTERPRET RESULT ---
label = "Parkinson" if prediction > 0.5 else "Healthy"
confidence = prediction if prediction > 0.5 else 1 - prediction

print(f"Prediction: {label} (Confidence: {confidence:.2f})")
```
